# comparison of CPU vs GPU

```python
import torch
import time

# Set the seed for reproducibility
torch.manual_seed(42)

# Define matrix dimensions
matrix_size = 10000  # Size of the square matrices

# Create random matrices
A = torch.rand(matrix_size, matrix_size)
B = torch.rand(matrix_size, matrix_size)

# Perform matrix multiplication on CPU
start_time_cpu = time.time()
C_cpu = torch.matmul(A, B)
end_time_cpu = time.time()
cpu_time = end_time_cpu - start_time_cpu

# Check if GPU is available
if torch.cuda.is_available():
    # Move matrices to GPU
    A_gpu = A.cuda()
    B_gpu = B.cuda()

    # Perform matrix multiplication on GPU
    start_time_gpu = time.time()
    C_gpu = torch.matmul(A_gpu, B_gpu)
    torch.cuda.synchronize()  # Ensure all GPU computations are finished
    end_time_gpu = time.time()
    gpu_time = end_time_gpu - start_time_gpu

    print(f"Matrix multiplication on CPU: {cpu_time:.6f} seconds")
    print(f"Matrix multiplication on GPU: {gpu_time:.6f} seconds")
else:
    print("GPU is not available on this system.")

# Results
if torch.cuda.is_available():
    print("\nSpeedup factor (CPU/GPU):", cpu_time / gpu_time)
else:
    print("\nOnly CPU computation was performed.")
```
